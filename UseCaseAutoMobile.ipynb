```python
# Input the data3
datalosses=pd.read_csv('automobile-losses.csv')
```

# PART 1

# 1. Join 3 tables using ID as key

```python
#merging dataspec and data risk
data12=pd.merge(dataspec, datarisk, on='ID') 
```

```python
#merging dataspec, data risk, adn data losses 
data123=pd.merge(data12, datalosses, on='ID')
data123
```

# 3. Identify missing value. If any, how will you handle it ? 

```python
#Checking is there any missing value by column?
data123.isnull().any()
```

```python
# How many the data-missing ? 
data123.isnull().sum()
```

```python
# Filling missing data
# numerical data fill with the mean
# categorical data fill with the modus
```

```python
data123.fillna(data123.mean(), inplace=True)
```

```python
data123['num-of-doors'].fillna(data123['num-of-doors'].mode()[0], inplace=True)
```
